# Dicionários em Python ( Chave e Valor )
Vimos listas ( nossa primeira estrutura de dados em Python ) e agora vamos entender como funcionam os dicionários.

## 1. O conceito
Se uma **Lista** é um armário com gavetas numeradas, um Dicionário em Python tem um funciamento muito parecido com um dicionário real: Quando você quer encontrar alguma coisa em um dicionário, você não busca pela "página 50", você busca pelo **termo (Chave)** para encontrar a **definição da palavra (Valor)**.

- **Lista**: Acesso por posição (``lista[0]``).

- **Dicionário**: Acesso por nome (``dicionario["nome"]``).

## 2. Sintaxe Básica

Dicionários usam chaves { } e pares de ``chave: valor``.
```py
    # Criando um dicionário de um aluno
    aluno = {
        "nome": "Maria",
        "curso": "Ciência da Computação",
        "semestre": 2
    }

    # Acessando um valor
    print(aluno["nome"]) 

    # Adicionando ou alterando dados
    aluno["nota"] = 9.5
    aluno["semestre"] = 3
```

## 3. Iterando em Dicionários (O ``for`` especial)

```py
    # 1. Percorrer apenas as chaves
    for chave in aluno:
        print(chave)

    # 2. Percorrer chaves e valores simultaneamente (Mais usado!)
    for chave, valor in aluno.items():
        print(f"{chave.capitalize()}: {valor}")
```

## 💡 Dicas
- **Chaves Únicas**: Explique que um dicionário não pode ter duas chaves iguais. Se você definir ``aluno["nome"] = "João"`` e depois ``aluno["nome"] = "Maria"``, o "João" é sobrescrito.

- **Performance**: Dicionários são extremamente rápidos para busca de dados, muito mais que listas, porque o Python usa uma técnica chamada *Hashing*.
