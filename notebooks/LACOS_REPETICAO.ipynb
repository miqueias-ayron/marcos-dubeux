# Laços ou Estruturas de Repetição
Imagine que você precise executar um mesmo código várias vezes. O que você faria? 

Para que você não precise ficar reescrevendo o mesmo código ou dando ``Ctrl+C , Ctrl+V`` inifitamente, existem os **Laços ou Estruturas de Repetição**. Essas estruturas viabilizam a execução de blocos de código em loop, evitando que você precise fazer isso manualmente. 

Existem duas principais estruturas de repetição: ``while`` e ``for``. Vejamos como isso funciona:

# Laço ``while`` em Python

## Definição
Em Python, o ``while``(enquanto) é uma estrutura de repetição. 
Usamos ``while`` quando não sabemos exatamente quantas vezes um bloco de código precisa ser repetido, mas sabemos a condição que deve ser mantida para ele continuar executando.

**analogia**: "Enquanto a xícara não estiver cheia, continue despejando café"

## Sintaxe
Em Python, fazemos assim:
```py 
    while condicao:
        # bloco de código que será repetido
        # atualização da condição (passo importante!)
```

## Exemplo Prático: Contador
Exemplo clássico para mostrar como o ``while`` precisa de uma variável de controle.

```py
    contador = 0

    while contador <= 5:
        print(contador)
        contador += 1 #se esquecer disso, o loop é infinito 
```

# Listas e o laço ``for`` em Python

## O que são Listas?
Se você sabe o que é uma variável, imagine uma variável como sendo um caixa que guarda um objeto ( basicamente qualquer coisa ). Uma **Lista** é um armário com com várias gavetas numeradas( índices ).

## Criando uma Lista
Lembra que uma variável é uma caixa que pode guardar qualquer coisa? Também vamos guardar listas dentro de variáveis!

```py
    transportes = ["carro", "moto", "onibus", "avião"]
```

Os elementos guardados dentro dessa lista podem ser acessados por seus índices, começando pelo ``0``. Então:

```py
  carro = transportes[0]
  onibus = transportes[2]    
```

### Que operações podemos fazer em uma lista?

![](imgs/listas.jpeg)

Além disso, também podemos fazer: 

- ``transportes[a:b]``: cria uma lista que vai do índice ``a`` ao ``b-1`` da lista transportes
- ``transportes[a::b]``: cria uma lista, começando no índice ``a``, pulando de ``b`` em ``b``.
- ``transportes[::]``: copia a lista
- ``transportes[::a]``: começa no índice ``0`` e pula de ``a`` em ``a``. 


### Os elementos de uma lista podem ser localizados de trás pra frente
Se queremos acessar o primeiro elemento da lista, usamos ``lista[0]`` mas se quisermos o **último** ( o primeiro de trás pra frente ), usamos ``lista[-1]``. Se quisermos o **penúltimo** ( o segundo de trás pra frente ), usamos ``lista[-2]``.

- ``transportes[::-1]``: copia a lista invertida
- ``transportes[::-2]``: inverte pulando de 2 em 2 começando pelo último.

# Laço ``for`` em Python
Diferente do ``while`` (que depende de uma condição booleana), o for em Python é um iterador. Ele percorre cada item de uma coleção do início ao fim.

## Sintaxe
```py
    for item in lista:
    # Faça algo com o item
```

## Exemplo prático:
```py
    compras = ["Arroz", "Feijão", "Café"]

    for produto in compras:
        print(f"Não esquecer de comprar: {produto}")
```

## A função ``range()``
Quando queremos repetir algo um número X de vezes (como um contador), usamos o ``range()``.
- ``range(5)`` gera: 0, 1, 2, 3, 4.

- **DICA**: Não altere uma lista enquanto você itera sobre ela.
